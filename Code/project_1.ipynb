```python
#Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import scipy.stats as sts
import gmaps
import gmaps.geojson_geometries
import time
from pprint import pprint

#Import API Keys
from config import (gkey, ckey)
```

```python
covid_data=pd.read_csv("Resources/us-counties.csv")
covid_data=covid_data.loc[covid_data["date"]== "2020-07-29"]
covid_data.reset_index(drop=True, inplace=True)
```

```python
covid_data["Lat"]=""
covid_data["Lng"]=""
covid_data
```

```python
#This cell takes a little over an hour to run
params={"key":gkey}
for index,row in covid_data.iterrows():
    base_url="https://maps.googleapis.com/maps/api/geocode/json"
    county=row["county"]
    state=row["state"]
    params["address"]=f"{county},{state}"
    result=requests.get(base_url, params=params)
    result=result.json()
    try:    
        covid_data.loc[index, "Lat"] = result["results"][0]["geometry"]["location"]["lat"]
        covid_data.loc[index, "Lng"] = result["results"][0]["geometry"]["location"]["lng"]
        time.sleep(1)
    except (IndexError, KeyError):
        print("results not found...skipping")
        time.sleep(1)
print("-"*30)
print("Complete")
```

```python
covid_data[["Lat","Lng"]]=covid_data[["Lat","Lng"]].apply(pd.to_numeric,axis=1)
covid_data.dropna(subset=["Lat"],inplace=True)
```

```python
covid_data
```

```python
gmaps.configure(api_key=gkey)
locations=covid_data[["Lat","Lng"]]
case_count=covid_data["cases"]
```

```python
fig=gmaps.figure()
heat_layer=gmaps.heatmap_layer(locations,weights=case_count,dissipating=False,point_radius=2,
                              max_intensity=case_count.max())
fig.add_layer(heat_layer)
fig
```
